```python
import transformers
import torch
import tqdm
```

```python
from transformers import pipeline, TextClassificationPipeline
from tqdm import tqdm
import time
```

```python
def time_text_classification(
    text_pipeline: TextClassificationPipeline,
    texts: list[str],
    batch_size: int = 1
) -> None:
  text_generator = (t for t in texts) # comprehensive
  pipeline_iterable = tqdm(
      text_pipeline(
          text_generator,
          batch_size=batch_size,
          truncation = True,
          max_length = 500
      ),
      total=len(texts)
  )
  for _ in pipeline_iterable:
    pass
```

```python
model_name = "cardiffnlp/twitter-roberta-base-sentiment-latest"
sentiment_pipeline_cpu = pipeline(model=model_name, device=-1)
sentiment_pipeline_gpu = pipeline(model=model_name, device=0)
```

```python
time_text_classification(sentiment_pipeline_cpu, reviews_list[:1000])
```

```python
time_text_classification(sentiment_pipeline_gpu, reviews_list[:1000])
```
